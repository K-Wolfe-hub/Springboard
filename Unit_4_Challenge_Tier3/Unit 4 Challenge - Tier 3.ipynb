#### 1.2.  Loading the data
Your data comes from the [London Datastore](https://data.london.gov.uk/): a free, open-source data-sharing portal for London-oriented datasets. 

```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

Don't be afraid to use StackOverflow for help  with this.

```python
properties = properties.transpose()
```

```python
properties = properties.reset_index()
```

```python
#pull each of the first row values into a list to provide column names.
col = []
for itm in properties.iloc[[0]].values:
    col.append(itm)
#fix the first two column names because they weren't actually in the first row
col[0][0] = 'Borough'
col[0][1] = 'BoroughCode'
#reassign the list of column names to the properties dataframe
properties.columns = col
#now that we have the columns in the right spot we should remove the first row that contains the column information
properties.drop(0,inplace=True)
```

```python
properties.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
#ok... so I got ahead of myself and did it a different way above... but here is how to do it the "right" way with rename...
properties.rename(columns={"Borough":"Borough","BoroughCode":"BCode"},inplace=True)
```

```python
properties.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
#get rid of the empty rows
properties.dropna(inplace=True)
#Getting rid of the extra Borough Code column
properties.drop(properties.columns[1],axis=1,inplace=True)
#melt the wide list of dates to allow us to process this data
proplong = pd.melt(properties,ignore_index=True,id_vars=[properties.columns[0]],var_name='Month')
#need to fix the Borough column index
proplong.rename(columns = {proplong.columns[0]:'Borough'},inplace=True)
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
proplong['value'] = proplong['value'].astype(float)
print(proplong.dtypes)
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# I used print(properties) to look at my full set of data.
# I somehow have 48 Boroughs in my dataset. It looks like there are sub-areas for index>33 and the first row is city of London
# need to get rid of the rows that are not Boroughs.
prop_just_boroughs = properties[1:33]
prop_just_boroughs.tail()
# I should reduce proplong to only include the 32 Boroughs as well. I'll start by pulling a list of the Boroughs I want to keep.
# this repeats the melt from above with just the 32 boroughs
proplong2 = pd.melt(prop_just_boroughs,ignore_index=True,id_vars=[prop_just_boroughs.columns[0]],var_name='Month')
#need to fix the Borough column index
proplong2.rename(columns = {proplong2.columns[0]:'Borough'},inplace=True)
proplong2['value'] = proplong2['value'].astype(float)
print(proplong2.dtypes)
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
brent = proplong2[proplong2.iloc[:,0] == 'Brent']
brent.plot(x='Month',y='value',kind='line',ylabel='Avg Price - $',title="Brent Pricing")
plt.show()                             
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column. 

To this end, you *could* apply a ***lambda function***. Your logic could work as follows:
1. look through the `Month` column
2. extract the year from each individual value in that column 
3. store that corresponding year as separate column. 

Whether you go ahead with this is up to you. Just so long as you answer our initial brief: which boroughs of London have seen the greatest house price increase, on average, over the past two decades? 

```python
#create a year column to reduce data points to something more manageable
proplong2['year'] = proplong2.apply(lambda row: row.Month.year, axis=1)
prop_yearly_ave = proplong2.groupby(by=['Borough','year']).mean()
prop_pivot = prop_yearly_ave.pivot_table(index='year',columns='Borough',values='value')
#Create a line plot of the all 32 districts
prop_pivot.plot(kind='line',ylabel='Avg Price - $',figsize=(15,10))
plt.show()
#Lets create a dataset that is the difference between 1998 to 2021
prop_98to21_diff = prop_pivot.loc[2021,:]-prop_pivot.loc[1998,:]
#Sort this to make a nicer looking plot
prop_98to21_diff.values.sort()
prop_98to21_diff.plot(kind='bar',x='Borough',sort_columns=True,figsize=(15,10),ylabel='1998 to 2021 Avg Price increase - $')
plt.show()
```

```python
#Create the suggested ratio of 1998 price to 2021 price. The lower this ratio, the higher the disparity between the two.
prop_98to21_ratio = prop_pivot.loc[1998,:]/prop_pivot.loc[2021,:]
prop_98to21_ratio.values.sort()
prop_98to21_ratio.plot(kind='bar',x='Borough',sort_columns=True,figsize=(15,10),ylabel='1998 Avg Price / 2021 Avg Price - $')
plt.show()
#Inerestingly although Westminster had the largest average price increase looking at the ratio of 1998 to 2021 prices Barking and Dagenham has seen a large change.
#just for fun lets also look a % increase (2021 price - 1998 price)/ 1998 price
prop_pct = (prop_pivot.loc[2021,:]-prop_pivot.loc[1998,:])/prop_pivot.loc[1998,:]*100
prop_pct.values.sort()
prop_pct.plot(kind='bar',x='Borough',sort_columns=True,figsize=(15,10),ylabel='Avg Price Percent Increase')
plt.show()
#Westminster seems to be the largest % increas at almost 600%!
print('Westminster 1998 Avg Price: $'+ str(int(prop_pivot.loc[1998,'Westminster'])))
print('Westminster 2021 Avg Price: $'+ str(int(prop_pivot.loc[2021,'Westminster'])))
print('Barking & Dagenham 1998 Avg Price: $'+ str(int(prop_pivot.loc[1998,'Barking & Dagenham'])))
print('Barking & Dagenham 2021 Avg Price: $'+ str(int(prop_pivot.loc[2021,'Barking & Dagenham'])))
```
